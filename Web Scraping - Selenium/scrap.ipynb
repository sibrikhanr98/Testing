```python
# using X_path find the element from html tags
from selenium.webdriver.common.by import By
all_matches_button = driver.find_element(By.XPATH,'//*[@id="content"]')

# click on the button
all_matches_button.click()

# to work with drop down list and selection
from selenium.webdriver.support.ui import Select
dropdown = Select(driver.find_element(By.ID,"country"))
dropdown.select_by_visible_text('Spain')

time.sleep(5)
# go to the tabel
matches = driver.find_elements(By.TAG_NAME,'tr')
#driver.quit()
```

```python
# create lists and assign each respective values to the page
# if we want to work with XPATH then we have to index it from 1
date = []
home_team = []
score = []
away_team = []

for match in matches:
    date.append(match.find_element(By.XPATH,'./td[1]').text)
    home_team.append(match.find_element(By.XPATH,'./td[2]').text)
    score.append(match.find_element(By.XPATH,'./td[3]').text)
    away_team.append(match.find_element(By.XPATH,'./td[4]').text)
driver.quit()
```

```python
# create dataframe

import pandas as pd

data = {'Date': date,
        'Home Team':home_team,
        'Score':score,
        'Away Team':away_team}
df = pd.DataFrame(data)
```

```python
# DataFrame to CSV
df.to_csv('FootBall_data.csv')
```

```python
# Youtube comments

from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
website = 'https://www.youtube.com/watch?v=aqXkKgkyk2I&ab_channel=SDET-QA'
driver.get(website)
time.sleep(7)
driver.maximize_window()

## SCROLL DOWN BY COMMENT SECTION

# Specify the Comment section
comment_section = driver.find_element(By.XPATH, '//*[@id="sections"]')
driver.execute_script("arguments[0].scrollIntoView(true);", comment_section)

# Wait for some time to load the comments
time.sleep(5)  # Adjust the wait time as needed

# Scroll down slowly to load comments
last_comment_count = 0
while True:
    # Scroll down to the comments section
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    
    # Wait for some time to load the comments
    time.sleep(2)
    
    # Break the loop if the last comment reached
    current_comment_count = len(driver.find_elements(By.XPATH, '//*[@id="content-text"]'))
    # If the number of loaded comments hasn't increased, break out of the loop
    if current_comment_count == last_comment_count:
        break

    # Update the last_comment_count
    last_comment_count = current_comment_count

# Find the comment elements using XPath
comments = driver.find_elements(By.XPATH, '//*[@id="content-text"]')

# Print the text of each comment
for i, comment in enumerate(comments):
    print(i + 1, ') ', comment.text, '\n')

#driver.quit()  # Close the browser after finishing
```
